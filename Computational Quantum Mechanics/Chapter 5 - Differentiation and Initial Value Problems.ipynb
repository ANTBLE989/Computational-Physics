**Example 5.3**

Use the forward Euler method to solve the initial value problem $y'(x) = x y(x)$, $y(0)=2$ between $0\leq x \leq 1$. 


```python
import numpy as np 
import matplotlib.pyplot as plt

# set x-grid size and spacing 

def euler_method(dx):
    x = np.arange(0,1+dx, dx)
    N = len(x)
    print(N)
    # create the y-grid 
    y = np.zeros((N))

    # set the initial condition 
    y[0] = 2 

    for i in range(0,N-1):
        # the Euler method 
        y[i+1] = y[i] + dx*(x[i]*y[i])
    return y


# dx = 0.1 
y1 = euler_method(0.01)
#y2 = euler_method(0.01)
#y3 = euler_method(0.001)
        
# exact solution 
dx = 0.01
x = np.arange(0,1+dx, dx)
ye= 2*np.exp((x**2)/2)

plt.plot(y1,'o-')
plt.plot(ye,'--')
```

**Example 5.9**

Use RK4 method for the DE problem in Example 5.9

```python
#!/usr/bin/env python3
import numpy as np

# define the ODE
def f(y,x):
    return x*y 


# set x-grid size and spacing 
dx = 0.01 
x = np.arange(0,1+dx,dx)
N = len(x)

# create the y-grid and initial conditions 
y = np.zeros((N))
y[0] = 2

# the RK4 method
for i in range(0,N-1):
    k1 = f(y[i],x[i])
    k2 = f(y[i]+0.5*k1*dx,x[i]+dx/2)
    k3 = f(y[i]+0.5*k2*dx,x[i]+dx/2)
    k4 = f(y[i]+k3*dx, x[i]+dx)
    
    y[i+1] = y[i] + dx*(k1 + 2*k2 + 2*k3 + k4)/6
```

```python
import matplotlib.pyplot as plt

# exact solution 
dx = 0.01
x = np.arange(0,1+dx, dx)
ye= 2*np.exp((x**2)/2)

plt.plot(y,'o')
plt.plot(ye,'--')
```
