## 3.5.1 The 1s Minimal STO-3G Basis Set

```python
x = np.linspace(-5,5,num=1000)
r = abs(x)
```

```python
def gaussian(alpha,r):
    return np.exp(-alpha*r*r)*(2*alpha/np.pi)**(3/4)

def slater(zeta,r):
    return np.exp(-zeta*r)*(((zeta)**3)/np.pi)**0.5
```

```python
sto_1g = gaussian(0.270950,r)
sto_2g = 0.678914*gaussian(0.151623,r)+0.430129*gaussian(0.851819,r)
sto_3g = 0.444635*gaussian(0.109818,r)+0.535328*gaussian(0.405771,r)+0.154329*gaussian(2.22766,r)
```

```python
plt.plot(r,slater(1,r),label="SLATER")
plt.plot(r,sto_1g,label="STO-1G",linestyle='dotted')
plt.plot(r,sto_2g,label="STO-2G",linestyle='dashdot')
plt.plot(r,sto_3g,label="STO-3G",linestyle='dashed')
plt.legend()
```

```python
plt.plot(x,slater(1,r),label="SLATER")
plt.plot(x,sto_1g,label="STO-1G",linestyle='dotted')
plt.plot(x,sto_2g,label="STO-2G",linestyle='dashdot')
plt.plot(x,sto_3g,label="STO-3G",linestyle='dashed')
plt.legend()
```

```python
# Radial distribution function 

slater_rad = (4*np.pi*r**2)*(slater(1,r)**2)
sto_1g_rad = (4*np.pi*r**2)*(sto_1g**2)
sto_2g_rad = (4*r**2)*(sto_2g**2)
```

```python
plt.plot(x,slater_rad,label="SLATER")
plt.plot(x,sto_1g_rad,label="STO-1G",linestyle='dashed')
#plt.plot(r,sto_2g_rad)
plt.legend()
```
